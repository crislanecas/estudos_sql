```python
import pandas as pd
from sqlalchemy import create_engine

# Carrega o arquivo CSV em um DataFrame.
df = pd.read_csv('/content/youtube_tratado.csv')

# Cria uma conexão com o SQLite.
engine = create_engine('sqlite:///youtube.db')

# Registra o DataFrame como uma tabela temporária no SQLite.
df.to_sql('youtube', con=engine, if_exists='replace', index=False)

# Carrega a extensão ipython-sql.
%load_ext sql

# Conecta-se ao banco de dados SQLite.
%sql sqlite:///youtube.db
```

```python
%%sql

PRAGMA table_info(youtube);
```

```python
%%sql

-- Seleciona todos os registros com palavra-chave TECNOLOGIA.
SELECT * FROM youtube
WHERE palavra_chave = 'TECNOLOGIA'
```

```python
%%sql

-- Qual foi o ano que teve mais visualizações?
SELECT ano, SUM(qtd_visualizacoes) AS total_visualizacoes
FROM youtube
GROUP BY ano
ORDER BY total_visualizacoes DESC;
```

```python
%%sql

-- Qual foi o mês com maior total de curtidas?
SELECT mes, SUM(qtd_curtidas) AS total_curtidas
FROM youtube
GROUP BY mes
ORDER BY total_curtidas DESC;
```

```python
%%sql

-- Qual são os top 3 assunto mais curtidos?
SELECT palavra_chave, SUM(qtd_curtidas) AS total_curtidas
FROM youtube
GROUP BY palavra_chave
ORDER BY total_curtidas DESC
LIMIT 3;
```

```python
%%sql

-- Qual são os top 3 assunto menos curtidos?
SELECT palavra_chave, SUM(qtd_curtidas) AS total_curtidas
FROM youtube
GROUP BY palavra_chave
ORDER BY total_curtidas ASC
LIMIT 3;
```

```python
%%sql

-- Qual a quantidade de vídeos que tem palavra-chave como 'Ciência de Dados'.
SELECT * FROM youtube
WHERE palavra_chave = 'CIÊNCIA DE DADOS';
```

```python
%%sql

-- Qual a média de curtidas por mês no ano de 2022?
SELECT mes, AVG(qtd_curtidas) AS media_curtidas
FROM youtube
WHERE ano = 2022
GROUP BY mes;
```

```python
%%sql

-- Qual vídeo pelo título teve maior quantidade de curtidas?
SELECT titulo, SUM(qtd_curtidas) AS total_curtidas
FROM youtube
GROUP BY titulo
ORDER BY total_curtidas DESC
LIMIT 3;
```

```python
%%sql

-- Qual vídeo pelo título teve menor quantidade de curtidas?
SELECT titulo, SUM(qtd_curtidas) AS total_curtidas
FROM youtube
GROUP BY titulo
ORDER BY total_curtidas ASC
LIMIT 3;
```
